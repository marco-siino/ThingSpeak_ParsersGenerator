```python
import json
from datetime import datetime
!wget 'https://raw.githubusercontent.com/marco-siino/ThingSpeak_ParsersGenerator/main/data.json'
with open('data.json') as f:
  data = json.load(f)
  output_data = []
  for feed in data['feeds']:
    timestamp = datetime.strptime(feed['created_at'], '%Y-%m-%dT%H:%M:%SZ')
    output_feed = {
    "sensorId": "San Diego - Estación Meteorológica",
    "timestamp": timestamp.isoformat(),
    "temperature": float(feed['field1']),
    "unit": "Celsius"
    }
  output_data.append(output_feed)
with open('output.json', 'w') as f:
  json.dump({"sensorId": "San Diego - Estación Meteorológica", "data": output_data}, f)
```
